## Подсчет числа уникальных элементов в группах

```python
data = pd.read_csv( 'teams_stats.csv' )
data.head()
```

```python
# проверим, что подсчет количества строк корректен

data.groupby( 'Year' ).count()['Tm'].head()
```

```python
# считаем количество уникальных элементов по годам

data.groupby( 'Year' ).nunique()['Tm'].head()
```

```python
# допольнительная разбивка по игрокам

data.groupby( ['Year', 'Player'] ).nunique()['Tm'].head()
```

## Фильтр, учитывающий для признака определенное число строк

```python
# данные отсюда https://www.kaggle.com/abcsds/highest-mountains
```

```python
# лимит количества строк, ниже которого не учитываем строки
LOWER_LIMIT = 3
```

```python
data = pd.read_csv( 'Mountains.csv' )
data.head()
```

```python
# считаем количество строк для каждой Parent mountain

parentMountainCount = data['Parent mountain'].value_counts()
parentMountainCount
```

```python
# фильтруем этот результат по нижней границе LOWER_LIMIT

filteredCount = parentMountainCount[ data['Parent mountain'].value_counts() > LOWER_LIMIT ]
filteredCount
```

```python
# переводим отфильтрованный список в лист

filteredList = filteredCount.index.tolist()
filteredList
```
